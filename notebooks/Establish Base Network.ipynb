### Base Network
Main goal here is to introduce my corpus to Networkx in Graph form. 
I want to draw the edges between poems that share the same word, while being able to subset each individual poem.

```python
from collections import defaultdict
from itertools import combinations

import networkx as nx
import nxviz as nv
import matplotlib.pyplot as plt

from gensim import corpora

% matplotlib notebook
```

So, right now I have a graph (corpus) where each node contains data specifying the document of origin, the term_id, and frequency of term in that document.
1. `vector_ind`: The document the node is associated with
2. `term_id`: The numeric representation of the term (can be cross-referenced in the dictionary); basically the node value. 
3. `freq_per_doc`: Frequency of `term_id` in `vector_ind`

My goal in this next section is to draw edges between nodes of different documents where the `term_id` is the same between documents. This edge will have a weight of 1.

```python
import networkx as nx
pickled_graph = G = nx.read_gpickle("../logic/test.gpickle")
```

```python
sub_g = list()

for n, d in pickled_graph.nodes(data=True):
    if d['vector_ind'] in [4, 5]:
        sub_g.append(n)
```

```python
subgraph = pickled_graph.subgraph(sub_g)
```

This is the result of the code written [here](../logic/create_edge_networks.py). 
For each poem, the same words have been connected across the poems, with a weight of one.

```python
import matplotlib as mpl

mpl.rcParams['figure.figsize'] = (20,20)

ap = nv.CircosPlot(subgraph, node_color='vector_ind', node_order='vector_ind', figsize=(8, 8))

ap.draw()
plt.title('Test')
plt.show();
```
