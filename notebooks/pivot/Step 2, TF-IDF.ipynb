```python
import pickle

from gensim.models import TfidfModel
from gensim.corpora import Dictionary
from gensim import corpora
```

```python
spa_dictionary = Dictionary.load('./spa.dict')
spa_corpus = corpora.MmCorpus('./spa.mm')
```

```python
spa_model = TfidfModel(spa_corpus)

spa_model.save('./spa-tf-idf.p')
```

```python
# take second element for sort
def take_second(elem):
    return elem[1]
```

```python
print("Top 5 terms in document `i` from SPA based on TF-IDF scoring")

for i in range(len(spa_corpus)):
    print(f"\n\n{i + 1}\n")
    
    vector = spa_model[spa_corpus[i]]
    vector.sort(key=take_second, reverse=True)
    
    counter = 0
    for term_id, tf_idf_score in vector:
        if counter > 4:
            break
            
        print(f"{spa_dictionary[term_id]} --> {tf_idf_score}")
        counter += 1
```
