```python
import networkx as nx
import matplotlib.pyplot as plt

% matplotlib inline
```

```python
g = nx.Graph()
g
```

```python
# Adding simple nodes
g.add_node('A')
g.add_node('B')
g.add_node('C')
g.add_node('D')

print(g.nodes)
```

```python

# adding edges
g.add_edge(u_of_edge='A', v_of_edge='B')
g.add_edge('A', 'D')
g.add_edge('B', 'D')
g.add_edge('B', 'C')
g.add_edge('C', 'D')

# Or, you could use:
# g.add_edges_from([('A','B'), ('A', 'D'), ...])

print(g.edges)
```

```python
# Adding in edge weights
edge_weights = {
    ('A', 'B'): 1,
    ('A', 'D'): 2,
    ('B', 'D'): 3,
    ('B', 'C'): 4,
    ('C', 'D'): 5,
}

nx.set_edge_attributes(g, edge_weights, 'weight')
print(g.edges(data=True))
```

```python
# You see all of that node's relationships, and the weights
# of the edges
g['A']
```

```python
# Accessing the data of A's relationshp to B
g['A']['B']
```

```python
# Accessing the weight of A's relationshp to B
g['A']['B']['weight']
```

```python
# since we have weights assigned to all edges.. 
[{(u,v): g[u][v]['weight']} for u, v in g.edges]
```

```python
# Simple graph!!
pos = nx.circular_layout(g)
nx.draw_networkx(g, pos)
```

```python
# Plotting the edges separately, and also by weight.
pos = nx.random_layout(g)
nx.draw_networkx_nodes(g, pos)

edge_widths = [d['weight'] for _,_,d in g.edges(data=True)]
nx.draw_networkx_edges(g, pos, width=edge_widths)
```

```python
# So, now that we have the basics down, we can define an 
# edge list programatically 

edge_list = [(x, y) for x in 'ABCDEFGH' for y in 'IJKLMNOP']
g2 = nx.from_edgelist(edge_list)

print(f'Nodes: {g2.nodes}')
print(f'Edges: {g2.edges}')
```

```python
pos = nx.circular_layout(g2)
nx.draw_networkx(g2, pos)
```
