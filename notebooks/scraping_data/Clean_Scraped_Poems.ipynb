There are a couple things about the poems that are off when I scraped them. 
First, there are blank lines in the files. 
I want to get rid of those.
Second, on every 5th line, there is a count. 
That also needs to be removed.

```python
for i in range(1, 247):
    lines = list()
    
    with open(f'{i}_poem.txt', 'r') as f:
        
        # removing the numbers at every 5th line in any given poem. 
        count = 0
        for line in f:
            if line.strip():
                count += 1
                
                if not count % 5:
                    line = line.replace(str(count), '')
                    
                lines.append(line.strip())
                
    # Either the first or second word in every poem is completely capitalized. 
    # This is especially annoying if you are trying to find names throughout the poems, making less meaning out of 
    # capitalized words. This process normalizes those. Ultimately I do a spot check and make sure they're good.
    first_line = lines[0].split(' ')
    first_two_words = first_line[:2]
    print(i, first_two_words)
    
    first_two_words = [word.lower() for word in first_two_words]
    first_two_words[0] = first_two_words[0].capitalize()
    
    # These are from spot-checking
    second_word_capitalize_list = [2, 16, 27, 35, 56, 66, 70, 84, 88, 95, 100, 102, 105, 107, 138, 160, 163, 166, 195, 197, 209, 223, 245]
    if i in second_word_capitalize_list:
        first_two_words[1] = first_two_words[1].capitalize()
    
    first_line[:2] = first_two_words
    lines[0] = ' '.join(first_line)
    print(lines[0])
    print()
           
    # Write out the cleaned poems.
    with open(f'{i}_cleaned_poem.txt', 'w') as f:
        for line in lines:
            f.write(line + '\n')
```
