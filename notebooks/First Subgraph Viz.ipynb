```python
from collections import defaultdict
from itertools import combinations

import networkx as nx
import nxviz as nv
import matplotlib.pyplot as plt

from gensim import corpora

% matplotlib inline
```

```python
subgraph = nx.read_gpickle("../logic/subgraph.gpickle")
```

```python
ap = nv.CircosPlot(
    subgraph, 
    node_color='vector_ind', 
    node_order='vector_ind', 
    node_grouping='vector_ind', 
    group_label_position='middle',
    group_label_color=True,
    figsize=(8, 8)
)

ap.draw()
plt.title('Network between Poem 4 and 5\n-1 represents common synonyms')

plt.savefig('first_full_network.pdf');
```
