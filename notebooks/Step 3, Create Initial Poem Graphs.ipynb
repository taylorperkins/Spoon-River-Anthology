# Step 3
## Create Base Graphs

I wanted to use networkx as a graph library to help visualize the progress as I go along.
This step is creating a super basic graph per poem using the info we already have.
No edges at this point, just nodes.

### Node structure
* `doc_id`: Points to the poem id
* `tf_idf`: TF-IDF score for a particular term within the poem that was determined in Step 2
* `term_id`: Numeric Id that makes back to the actual word. The hash dictionary is being stored in the `spa.dict`
* `type`: SPA. This will allow us to easily differenciate the poem nodes from their related synonyms.

```python
import networkx as nx
from gensim.models import TfidfModel
from gensim.corpora import Dictionary
from gensim import corpora
```

```python
tf_idf_model = TfidfModel.load('../data/spa-tf-idf.p')
spa_corpus = corpora.MmCorpus('../data/spa.mm')

total_nodes = 0
for ind, document_vector in enumerate(spa_corpus):
    if ind % 1 == 0:
        print(ind)
    
    tf_idf_vector = tf_idf_model[document_vector]
    
    nodes = list()
    for term_id, tf_idf_score in tf_idf_vector:
        nodes.append((
            total_nodes, 
            {
                'doc_id': ind,
                'tf_idf': tf_idf_score,
                'term_id': term_id,
                'type': 'SPA'
            }
        ))
        
        total_nodes += 1
    
    poem_graph = nx.Graph()
    poem_graph.add_nodes_from(nodes)
    
    nx.write_gpickle(poem_graph, f"../data/base_graphs/{ind}.gpickle")
```

### No edges, just nodes
You can see all of the node attributes from this view.
This happens to be the longest poem (by a longshot).
Don't get overwhelmed by the view.
