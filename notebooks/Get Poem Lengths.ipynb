```python
import pickle


poem_lengths = dict()

base_dir = '../data/subgraphs/'

for ind, dir_name in enumerate(os.listdir(base_dir)):
    print(f"Finding counts for {ind}")
    
    poem_dir = base_dir + dir_name + '/'
    
    for j_ind, file in enumerate(os.listdir(poem_dir)):
        print(f"Reading subgraph for {j_ind}")
        
        file_path = poem_dir + file 
        second_poem = file.split('.')[0]
        
        subgraph = nx.read_gpickle(file_path)
        
        word_count = 0

        for node, data in subgraph.nodes(data=True):
            if data['type'] == 'SYN':
                word_count += 1

            else:
                word_count += data['freq_per_doc']
                
        poem_lengths[(dir_name, file)] = int(word_count)
        
        
pickle.dump(poem_lengths, open('poem_lengths.p', 'wb'))
```
